План анализа:


1) загрузка и получение общей информации о датасетах;

2) работа с пропусками, дубликатами, типами данных (выявление, обоснование сохранения/удаления, удаление), созданием новых столбцов;

6) визуализация важности входных признаков, рекомендации для заказчика;

7) прогноз статуса стартапа;

8) выделение признаков, влияющих на продолжение стартапом своей деятельности;

9) итоговые выводы.

Мы создали в тренировочном (по образцу тестового) датасете столбец lifetime через вычисление разницы между датой закрытия и датой создания (или первого финансирования, если дата создания позже последнего финансирования). Из этого следует, что дату закрытия можно определить по сумме lifetime  и даты основания (а если дата основания позже даты последнего финансирования, то используем дату певого финансирования). Затем из даты закрытия (уже для тестового датасета) вычтем дату последнего финансирования и получим искомое количество дней между датой закрытия (для тестового датасета!) и датой последнего финансирования в нем. Отказалась от этой идеи, т.к., возможно, это будет дополнительной подсказкой модели для предсказания статуса.
</div>

Дополним тренировочный датасет столбцом с количеством дней между датой первого финансирования и датой создания компании. Исходим из того, что нам важно, когда компания получает первые деньги. Если дата создания компании позже даты последнего финансирования (возможно, из-за того, что не совсем хорошо идут дела в компании), то вычислять будем не разность между датами первого финансирования и создания, а между датами создания и первого финансирования. Считаем, что если компания создана (юридически) позднее первого финансирования, то "встала на ноги" уже после первого финансирования и решила официально оформить свой статус.

```python
train['days_first_money_after_open'] = (train['first_funding_at'] - train['founded_at']).dt.days
train.loc[train['days_first_money_after_open'] < 0, 'days_first_money_after_open'] = (train['founded_at'] - train['first_funding_at']).dt.days
train.loc[train['name'] == "Kraftwurx"]
```

Далее начнем работу с моделями для определения лучшей модели. Сначала используем повышенное число признаков, затем сократим их для проверки работы модели.

```python
RANDOM_STATE = 42
TEST_SIZE=0.25

X_train, X_test, y_train, y_test = train_test_split(
    train_without_name_new.drop([
                                       'status',
                                       'category_list',
                                       'country_code',
                                       'founded_at',
                                       'first_funding_at',
                                       'last_funding_at',
                                       'state_code',
                                       'region',
                                       'city',
                                       'funding_total_usd'
], axis=1),
    train_without_name_new['status'],
    test_size = TEST_SIZE, 
    random_state = RANDOM_STATE,
    stratify = train_without_name_new['status'])
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(y_train)
y_test = label_encoder.transform(y_test)


# создаём списки с названиями признаков
ohe_columns = ['general_category', 'geographical_zone', 'founded_at_ym', 'first_funding_at_ym', 'last_funding_at_ym']
#ord_columns = ['supervisor_evaluation']
num_columns = ['funding_rounds', 'funding_per_round', 'lifetime', 'days_first_money_after_open']

# создаём пайплайн для подготовки признаков из списка ohe_columns: заполнение пропусков и OHE-кодирование
# SimpleImputer + OHE
ohe_pipe = Pipeline(
    [('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore'))
     
    ]
    )

# создаём общий пайплайн для подготовки данных
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     #('ord', ord_pipe, ord_columns),
     ('num', StandardScaler(), num_columns)
    ], 
    remainder='passthrough'
)

# создаём итоговый пайплайн: подготовка данных и модель
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])

param_grid = [
    # словарь для модели DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 5),
        'models__max_features': range(2, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']  
    },
    
    # словарь для модели KNeighborsClassifier() 
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']   
    },

    # словарь для модели SVC() 
    {
        'models': [SVC(probability=True, random_state=RANDOM_STATE)],
        'models__kernel': ['linear', 'rbf', 'poly'],
        'models__degree': [2, 3, 4],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']   
    },
    
    # словарь для модели LogisticRegression()
    {
        'models': [LogisticRegression(
            random_state=RANDOM_STATE, 
            solver='liblinear', 
            penalty='l1'
        )],
        'models__C': range(1, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']  
    },
    
    # словарь для модели RandomForest()
    {
        'models': [RandomForestClassifier()],
        'models__n_estimators': range (10, 51, 10),
        'models__max_depth': range (1,13, 2),
        'models__min_samples_leaf': range (1,15),
        'models__min_samples_split': range (2,15,2) ,
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler()]
    }
    
]
randomized_search = RandomizedSearchCV(
    pipe_final, 
    param_grid, 
    cv=5,
    scoring='f1', # используем данную метрику по заданию
    random_state=RANDOM_STATE,
    n_jobs=-1
)
randomized_search.fit(X_train, y_train)

print('Лучшая модель и её параметры:\n\n', randomized_search.best_estimator_)
print ('Метрика лучшей модели на тренировочной выборке:', randomized_search.best_score_)
probabilities = randomized_search.predict(X_test)

print('Метрика на тестовой выборке:', f1_score(y_test, probabilities))
```

Также для OneHotEncoder сохранены пераметры drop='first', handle_unknown='ignore' (а не error), потому что после обновления sklearn до последней версии конфликт между этими параметрами сохранился.

Далее на различных пороговых значениях (превышающих установленные по умолчанию 0,5) посмотрим на величину метрики.
